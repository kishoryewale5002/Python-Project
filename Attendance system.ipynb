```python
employees = {}

# Function to add employee
def add_employee():
    emp_id = input("Enter Employee ID: ")
    name = input("Enter Employee Name: ")
    daily_salary = float(input("Enter Daily Salary: "))
    
    employees[emp_id] = {
        "name": name,
        "daily_salary": daily_salary,
        "present_days": 0
    }
    
    print("Employee added successfully!\n")


# Function to mark attendance
def mark_attendance():
    emp_id = input("Enter Employee ID: ")
    
    if emp_id in employees:
        days = int(input("Enter number of days present: "))
        employees[emp_id]["present_days"] = days
        print("Attendance updated!\n")
    else:
        print("Employee not found!\n")


# Function to calculate salary
def calculate_salary():
    emp_id = input("Enter Employee ID: ")
    
    if emp_id in employees:
        daily = employees[emp_id]["daily_salary"]
        days = employees[emp_id]["present_days"]
        salary = daily * days
        
        print("Employee Name:", employees[emp_id]["name"])
        print("Days Present:", days)
        print("Total Salary:", salary)
        print()
    else:
        print("Employee not found!\n")


# Function to display report
def display_report():
    print("\nEmployee Report")
    print("----------------------------")
    
    for emp_id, data in employees.items():
        salary = data["daily_salary"] * data["present_days"]
        
        print("ID:", emp_id)
        print("Name:", data["name"])
        print("Days Present:", data["present_days"])
        print("Total Salary:", salary)
        print("----------------------------")


# Main program
while True:
    
    print("EMPLOYEE ATTENDANCE & SALARY ANALYZER")
    print("1. Add Employee")
    print("2. Mark Attendance")
    print("3. Calculate Salary")
    print("4. Display Report")
    print("5. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        add_employee()
        
    elif choice == "2":
        mark_attendance()
        
    elif choice == "3":
        calculate_salary()
        
    elif choice == "4":
        display_report()
        
    elif choice == "5":
        print("Exiting Program...")
        break
        
    else:
        print("Invalid Choice!\n")
```
